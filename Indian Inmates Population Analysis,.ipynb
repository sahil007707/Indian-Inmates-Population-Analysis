## Data Handling & Cleaning

```python
df.isna().sum()
df.duplicated().sum()
```

```python
df=df.rename(columns={"State/UT":"States"})
```

## EDA & Visualization

```python
df.head()
```

```python
classifi_p=df.groupby("States")[["Male","Female"]].sum().reset_index()
classifi_p
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # Creating a figure with 2 subplots side by side
sns.set_style("whitegrid")

# First subplot
sns.barplot(data=classifi_p, x="States", y="Male", ax=axes[0])
axes[0].set_title("Male Inmates Distribution")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)

# Second subplot
sns.barplot(data=classifi_p, x="States", y="Female", ax=axes[1])
axes[1].set_title("Female Inmates Distribution")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

## Total Inmate Population Per States

```python
df.head()
```

```python
classifi_t=df.groupby("States")["Total"].sum().reset_index()
plt.figure(figsize=(15,6))
sns.barplot(data=classifi_t,x="States",y="Total",color="darkred")
plt.xticks(rotation=90)
plt.show()
```
